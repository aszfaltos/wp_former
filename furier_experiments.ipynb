```python
from data_handling import data_loader
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.dates as mdates
import math
```

```python
df = data_loader.load_mavir_data('data/mavir_data/mavir.csv')
```

```python
# standardize power
df['Power'] = (df['Power'] - df['Power'].mean()) / df['Power'].std()
# convert time to datetime
df['Time'] = pd.to_datetime(df['Time'])

# chose the first n datapoints as a sample
SAMPLE_SIZE = 10
SAMPLE_START = 0
sample = df.iloc[SAMPLE_START:SAMPLE_START+SAMPLE_SIZE, :]
```

```python
sp = np.fft.fft(sample['Power'])
freq = np.fft.fftfreq(sample['Power'].shape[-1])
freq
```

```python
plt.rcParams['figure.figsize'] = [14, 15]

plt.plot(freq, sp.real, '--')#, freq, sp.imag, '--')

plt.show()
```
